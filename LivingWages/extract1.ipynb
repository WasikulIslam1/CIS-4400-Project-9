```python


from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
```

```python


url = "https://files.epi.org/uploads/fbc_livingwage_data_2024.xlsx"


response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Write the content of the response to a local file
    with open("fbc_livingwage_data_2024.xlsx", "wb") as file:
        file.write(response.content)
    print("File downloaded successfully.")

    # Read the Excel file into a pandas DataFrame
    df = pd.read_excel("fbc_livingwage_data_2024.xlsx")

    # Save DataFrame to CSV
    df.to_csv("fbc_livingwage_data_2024.csv", index=False)

    print("Excel file converted to CSV successfully.")
else:
    print("Failed to download the file.")
```
