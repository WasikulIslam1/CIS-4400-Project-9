```python
import os
import json
import pandas as pd

# Get the current working directory
json_dir = os.getcwd()

# Initialize an empty list to store data from each JSON file
data_list = []

# Check if the directory exists
if not os.path.exists(json_dir):
    print("Directory does not exist:", json_dir)
else:
    # Iterate over each file in the directory
    for filename in os.listdir(json_dir):
        if filename.endswith('.json'):  # Ensure it's a JSON file
            file_path = os.path.join(json_dir, filename)
            try:
                with open(file_path) as json_file:
                    data = json.load(json_file)
                    data_list.append(data)
            except Exception as e:
                print(f"Error reading file {file_path}: {e}")

    # Convert the list of dictionaries into a pandas DataFrame
    df = pd.DataFrame(data_list)

    # Now you have a DataFrame containing data from all JSON files in the directory
    print(df)
```

```python
import os
import json
from azure.storage.blob import BlobServiceClient

config_file_path = 'config.json'

with open(config_file_path, 'r') as config_file:
    config = json.load(config_file)

CONNECTION_STRING_AZURE_STORAGE = config["connectionString"]
container_name = "mitraw"
base_folder = "alljsonfiles"  # Base folder name in Azure Storage

# Function to upload a file to Azure Blob Storage
def upload_file_to_blob(file_path, blob_name):
    blob_service_client = BlobServiceClient.from_connection_string(CONNECTION_STRING_AZURE_STORAGE)
    container_client = blob_service_client.get_container_client(container_name)
    blob_client = container_client.get_blob_client(blob_name)
    with open(file_path, "rb") as data:
        blob_client.upload_blob(data)

# Get the current working directory
json_dir = os.getcwd()

# Check if the directory exists
if not os.path.exists(json_dir):
    print("Directory does not exist:", json_dir)
else:
    # Iterate over each file in the directory
    for root, dirs, files in os.walk(json_dir):
        for file in files:
            if file.endswith('.json'):  # Ensure it's a JSON file
                file_path = os.path.join(root, file)
                # Get relative path (without the current working directory)
                relative_path = os.path.relpath(file_path, json_dir)
                # Replace backslashes with forward slashes (for Azure Blob Storage)
                blob_name = os.path.join(base_folder, relative_path).replace("\\", "/")
                # Upload the file to Azure Blob Storage
                upload_file_to_blob(file_path, blob_name)
                print(f"Uploadeds {file} to Azure Storage as {blob_name}")
```
