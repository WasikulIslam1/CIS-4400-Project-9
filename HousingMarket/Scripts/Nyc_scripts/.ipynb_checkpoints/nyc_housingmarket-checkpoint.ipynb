```python
# Import libraries
from io import StringIO
import numpy as np
import pandas as pd
import json
import requests
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
```

```python
#URLs for each borough's Excel file

URL_manhattan = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_manhattan.xlsx"
URL_bronx = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_bronx.xlsx"
URL_brooklyn = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_brooklyn.xlsx"
URL_queens = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_queens.xlsx"
URL_statenisland = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_statenisland.xlsx"

# Loading each Excel file into a separate dataframe
df_manhattan = pd.read_excel(URL_manhattan)
df_bronx = pd.read_excel(URL_bronx)
df_brooklyn = pd.read_excel(URL_brooklyn)
df_queens = pd.read_excel(URL_queens)
df_statenisland = pd.read_excel(URL_statenisland)

# Display information about each dataframe
print("Manhattan DataFrame Info:")
print(df_manhattan.info())
print(df_manhattan.head())

print("Bronx DataFrame Info:")
print(df_bronx.info())
print(df_bronx.head())

print("Brooklyn DataFrame Info:")
print(df_brooklyn.info())
print(df_brooklyn.head())

print("Queens DataFrame Info:")
print(df_queens.info())
print(df_queens.head())

print("Staten Island DataFrame Info:")
print(df_statenisland.info())
print(df_statenisland.head())
```
