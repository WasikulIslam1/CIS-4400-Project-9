```python
# Define URLs for each borough's sales data
URL_manhattan = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_manhattan.xlsx"
URL_bronx = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_bronx.xlsx"
URL_brooklyn = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_brooklyn.xlsx"
URL_queens = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_queens.xlsx"
URL_statenisland = "https://www.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_statenisland.xlsx"

# Load data from each URL into a dataframe, set the header from the fourth row (index 3), and drop rows with any NaN values
df_manhattan = pd.read_excel(URL_manhattan, header=4)
df_bronx = pd.read_excel(URL_bronx, header=4)
df_brooklyn = pd.read_excel(URL_brooklyn, header=4)
df_queens = pd.read_excel(URL_queens, header=4)
df_statenisland = pd.read_excel(URL_statenisland, header=4)

nyc_df = pd.concat([df_manhattan, df_bronx, df_brooklyn, df_queens, df_statenisland], ignore_index=True)
nyc_df.info()
```

```python
# Specify the path to your JSON configuration file
config_file_path = 'config.json'

# Load the JSON configuration file
with open(config_file_path, 'r') as config_file:
    config = json.load(config_file)
    
CONNECTION_STRING_AZURE_STORAGE = 'connectionString'
CONTAINER_AZURE = 'nychousingsales'
blob_name = 'nychousingsales'

#Convert DATAFRAME to csv
output= StringIO()
nyc_df.to_csv(output, index=False)
data = output.getvalue()
output.close()

# Create the BlobServiceClient object
blob_service_client = BlobServiceClient.from_connection_string(CONNECTION_STRING_AZURE_STORAGE)

# Get a blob client using the container name and blob name
blob_client = blob_service_client.get_blob_client(container=CONTAINER_AZURE, blob=blob_name)

# Upload the CSV data
blob_client.upload_blob(data, overwrite=True)

print(f"Uploaded {blob_name} to Azure Blob Storage in container {CONTAINER_AZURE}.")
```
