```python
from bs4 import BeautifulSoup
import requests
import csv

state_abbreviations = [
    "AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA",
    "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
    "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
    "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
    "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"
]

city_data = []
for state in state_abbreviations:
    url = f"https://spotcrime.com/{state}"
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    header = soup.find("h1", class_="state-page__page-section__header")
    div_sibling = header.find_next_sibling("div")
    city_containers = div_sibling.find_all("div", class_="state-page__city-container")
    for container in city_containers:
        city_link = container.find("a", class_="state-page__city-btn")
        city_name = city_link.text.strip()
        city_data.append(city_name.split(", "))

city_dict = {}

for city_string in city_data:
    # Check if there are at least two elements after splitting
    if len(city_string) >= 2:
        state_code = city_string[1]
        city_name = city_string[0]
        
        if city_name == "Murrysville":
            city_dict["PA"].append("Murrysville")
        # If state code exists as a key, append city name to its list
        if state_code in city_dict:
            city_dict[state_code].append(city_name)
        else:
            # If state code is not a key, create a new key-value pair
            city_dict[state_code] = [city_name]
del city_dict["Export"]

all_data = []
# Assuming you have city_dict populated with city names for each state
for state, cities in city_dict.items():
    for city in cities:
        url = f"https://spotcrime.com/{state}/{city}"
        response = requests.get(url)
        data = response.text

        # Scraping trends data
        url_trends = f"https://spotcrime.com/{state}/{city}/trends"
        response_trends = requests.get(url_trends)
        data_trends = response_trends.text

        # Parsing HTML data
        soup = BeautifulSoup(data, 'html.parser')
        soup_trends = BeautifulSoup(data_trends, 'html.parser')

        # Extracting trends data
        rows_trends = soup_trends.find_all("td")
        list_of_trends = [row.text.strip() for row in rows_trends]

        # Store the data in a list
        city_data = [state] + list_of_trends
        all_data.append(city_data)

# Write the data to a CSV file
with open('crime_data.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    # Write header row
    writer.writerow(["State", "City", "Thefts Last Week", "Thefts Last 30 Days", "Thefts Previous 30 Days", "Theft Last Six Months", "Robbery Last Week", "Robbery Last 30 Days", "Robbery Last Six Months", "Burglary Last Week", "Burglary Last 30 Days", "Burglary Previous 30 Days", "Burglary Last Six Months", "Vandalism Last Week", "Vandalism Last 30 Days", "Vandalism Previous 30 Days", "Vandalism Last Six Months", "Shooting Last Week", "Shooting Last 30 Days", "Shooting Previous 30 Days", "Shooting Last Six Months", "Arson Last Week", "Arson Last 30 Days", "Arson Previous 30 Days", "Arson Last Six Months", "Arrest Last Week", "Arrest Last 30 Days", "Arrest Previous 30 Days", "Arrest Last Six Months", "Assault Last Week", "Assault Last 30 Days", "Assault Previous 30 Days", "Assault Last Six Months", "Other Last Week" ,"Other Last 30 Days", "Other Previous 30 Days", "Other Last Six Months"] + ["Trends Data"] * len(list_of_trends))
    # Write city data
    writer.writerows(all_data)

print("CSV file 'crime_data.csv' created successfully.")
```
